```python
import pandas as pd
import numpy as np

data = {
    'Product' : ['Laptop', 'Camera', 'Mobiles', 'Smartwatch', 'Tablet', 'Headphones'],
    'Price' : [125000, 50000, 80000, 40000, 30000, 10000 ],
    'Units_sold' : [100, 50, 200, 60, 40, 75]
    }

df = pd.DataFrame(data)
print(df)
```

```python
#Data binning

bins1 =  [0, 50000, 100000, 150000]
labels1 = ['Low', 'Medium', 'High']

df['Price_Category'] = pd.cut(df['Price'], bins=bins1, labels=labels1)
print(df)
```

```python
#Binning units sold

bins2 = [0, 50, 100, 200]
labels2 = ['Low sold', 'Moderate sold', 'High sold']

df['Units_Category'] = pd.cut(df['Units_sold'], bins=bins2, labels=labels2)
print(df)
```

###Data Formatting

```python
#Convert data types

print(df.dtypes)

df['Price'] = df['Price'].astype(float)

print(df.dtypes)
```

```python
df
```

```python
#Convert Text to uppercase

df['Product'] = df['Product'].str.upper()
print(df)
```

```python
#Rounding values

df['Price'] = df['Price'].round(2)
print(df)
```

###Additional operators

```python
#Sort data

sorted_df = df.sort_values(by='Price')
print(sorted_df)
```

```python
#Display unique values

print(df['Price_Category'].unique())
```

```python
import pandas as pd
import numpy as np

data = {
    'Order_ID' : [101, 102, 103, 104, 105, 106, 107 , 108],
    '0rder_Value' : [250, 300, 550, 400, 380, 700, 675, 900],
    'Delivery_time' : [30, 15, 37, 45, 55, 20, 22, 48],
    'Distance_KM' : [5,2,6,7,4,8,3,9]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Binning delivery time

bins1 = [0,25,40,60]
labels1 = ['Fast', 'Average', 'Slow']

df['Delivery_speed'] = pd.cut(df['Delivery_time'], bins=bins1, labels=labels1)
print(df)
```
