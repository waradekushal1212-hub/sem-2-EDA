# Operations in Sets

```python
# Creation of set

f_set = {'sandals', 'flip', 'chapal', 'shoes', 'sandals', 'formal shoes'}
print(f_set)
```

```python
# Duplicates not allowed
f_set = {'boots', 'flip flops', 'sneakers', 'shoes', 'boots', 'boots'}
print(f_set)
```

```python
# The values True and 1 are considered same value in sets
# same is applicable for 0 and false

f_set = {'correct', 'wrong', 'incorrect', True, 1 , 2}
print(f_set)
```

```python
# A set with integers, Strings, and Boolean Values

s_set = {"ENTC", 2025, True, 1 , False, 18.7}
print(s_set)
```

```python
# To check Membership

f_set = {'boots', 'flip flops', 'sneakers', 'shoes', 'sandals', 'formal shoes'}
print("boots" in f_set)
print(True in f_set)
```

```python
# Once a set is created, we can't replace its items but we can add new items

f_set = {'boots', 'flip flops', 'sneakers', 'shoes', 'sandals', 'formal shoes'}
f_set.add("slippers")
print(f_set)
```

```python
# Remove items from the set

f_set.remove("shoes")
print(f_set)
```

```python
# Set Operations

first = {5,10,15,20,25,30}
second = {"Lehalyaa", "Aditi", 10, 30 , 3}

print("The union of two sets is: ", first | second)
print("The intersection of two sets is: ", first & second)
print("The difference of two sets is: ", first - second)
print("The Symmetric Difference of two sets is: ", first ^ second)
```

```python
# Frozen Set

frozen = frozenset({"boots", "flip flops", "sneakers", "shoes", "sandals", "formal shoes"})
print(frozen)
print(type(frozen))
```

```python
# Adding or removal of elements are not allowed in frozen set

print("Values can be added to frozen set by union \n", frozen | {"slippers"})
frozen.add("slipper")
frozen.remove("boots")
```

```python
''' Problem Statement - 4 - A Dictionary stores student names and marks. Write a
Python program to find the student who scored the highest marks'''

marks = {"KUSHAL" : 90, "Aditi" : 95, "Gayatri" :90, "kushal" : 93, "Arjit" : 98, }
max_stu = max(marks, key = marks.get)
print("Student with Highest marks: ", max_stu)
print("The highest marks are: ", marks[max_stu])
```
