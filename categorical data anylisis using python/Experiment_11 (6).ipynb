```python
import pandas as pd

data = {

'OrderID' : [ '01', '02', '03', '04', '05', '06', '07', '08', '09', '010'],

'Category' : [ 'Electronics', 'Clothing', 'Cosmetics', 'Grocery', 'Clothing', 'Electronics', 'Cosmetics', 'Grocery', 'Grocery', 'Clothing'],

'Payment' : ['UPI', 'COD', 'UPI', 'Netbanking', 'UPI', "COD", "UPI", "UPI", "Netbanking", "UPI"],

'Delivery' : ['Standard', 'Standard', 'Standard', 'Express', 'Standard', 'Express', 'Standard', 'Express', 'Express', 'Standard'],

'CustomerType' : ['New', 'New', 'Returning', 'Returning', 'New', 'New', 'Returning', 'Returning', 'New', 'Returning']
}

df = pd.DataFrame(data)

print(df)
```

```python
#Frequency count (Count how many orders in each category)

print(df['Category'].value_counts())
```

```python
#Unique values

print(df['Category'].unique())
```

```python
#Number of unique values (will give the count - number of varieties)

print(df['Category'].nunique())
```

```python
#Frequency of payment methods

print(df['Payment'].value_counts())
```

```python
#Cross Tabulation (category vs payment method)

print(pd.crosstab(df['Category'], df['Payment']))
```

```python
#Cross tabulaion (Delivey vs customer type)

print(pd.crosstab(df['Delivery'], df['CustomerType']))
```

```python
#Percentage distribution

print(df['Category'].value_counts(normalize=True)*100)
```

```python
#Filtering data (eg - orders from electronics category)

electronics_orders = df[df['Category']=='Electronics']
print(electronics_orders)
```

```python
#Filtering data (Grocery)

grocery_orders = df[df['Category']=='Grocery']
print(grocery_orders)
```

```python
#Grouping (orders by category and payment method)

print(df.groupby('Category')['Payment'].value_counts())
```

```python
#Sorting data

sorted_df = df.sort_values(by='Category')
print(sorted_df)
```
