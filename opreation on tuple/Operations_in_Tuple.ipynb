```python
## Python Tuples
# Access Tuple Items

brands = ("apple", "dnm", "Rolex", "rado", "Titan", "Sonata")
print(brands[3])
```

```python
# Tuple is Immutable

brands = ("prada", "Lifestyle", "Rolex", "Casio", "Titan", "Sonata")
brands[2] = "Boat"
print(brands)
```

```python
# Addition of Tuples

t1 = (1, 3, 7, 9)
t2 = (2, 4, 6, 8)
t3 = t1 + t2
print("t1: ", t1)
print("t2: ", t2)
print("t3: ", t3)
```

```python
#Address Verification after Appending new item in the Tuple

brands = ("prada", "Lifestyle", "Rolex", "Casio", "Titan", "Sonata")
print("Before: ", id(brands), "\n")
new = ("Ferrari",)
brands = brands + new
print("After: ", id(brands),"\n")
```

```python
#Address Verification after Appending new item in the List

brands = ["prada", "Lifestyle", "Rolex", "Casio", "Titan", "Sonata"]
print("Before: ", id(brands), "\n")
new = ["Ferrari"]
brands += new
print("After: ", id(brands),"\n")
```
