# Problem Statements

```python
# student marks

phy = int(input("Enter your Physics marks: "))
chem = int(input("Enter your Chemistry marks: "))
bio = int(input("Enter your Biology marks:"))
eng = int(input("Enter your English marks: "))
math = int(input("Enter your Mathematics marks: "))
sum = phy + chem + bio + eng + math
print("Your total marks are: ", sum)
print("Your Average mark is: ", sum / 5 )
```

```python
# Area of Circle

r = float(input("Enter the radius of the circle: "))
area = 3.14 * r * r
print("The area of the circle of the given radius is: ", area)
```

```python
# Area of a Rectangle

l  = float(input("Enter the length of the rectangle: "))
b = float(input("Enter the breadth of the rectangle: "))
area = l * b
print("The area of the rectangle is: ", area)
```

```python
# Perimeter of Rectangle

l  = float(input("Enter the length of the rectangle: "))
b = float(input("Enter the breadth of the rectangle: "))
perimeter = 2 * (l + b)
print("The perimeter of the rectangle is: ", perimeter)
```

```python
# Simple Calculator

n1 = float(input("Enter the first number: "))
n2 = float(input("Enter the second number: "))
print("Enter \n + for Addition \n - for Subtraction \n * for Multiplication \n / for Division")
op = input("Enter the operator:")
if(op == '+'):
  print("The sum is: ", n1 + n2)
elif(op == '-'):
  print("The difference is: ", n1 - n2)
elif(op == '*'):
  print("The product is: ", n1 * n2)
elif(op == '/'):
  print("The quotient is: ", n1 / n2)
else:
  print("Invalid Operator")
```
