## 14A -Data Normalization

```python
import pandas as pd
import numpy as np

data = {
    'Product' : ['Laptop', 'Camera', 'Mobiles', 'Smartwatch', 'Tablet', 'Headphones'],
    'Price' : [125000, 50000, 80000, 40000, 30000, 10000 ],
    'Units_sold' : [100, 50, 200, 60, 40, 75],
    'Discount' : [10, 8, 18, 12, 20, 15]
    }

df = pd.DataFrame(data)
print("Original datasheet")
print(df)
```

###Min-Max normalization

```python
#Min-Max normalization for price :

df['Price_MinMax'] = (df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())

print("\nMix-Max Normalized Price")
print(df[['Product','Price','Price_MinMax']])
```

```python
#Min-Max normalization for discounts :

df['Discount_MinMax'] = (df['Discount']-df['Discount'].min())/(df['Discount'].max()-df['Discount'].min())

print("\nMix-Max Normalized Discount")
print(df[['Product','Discount','Discount_MinMax']])
```

```python
#Min-Max normalization for units sold :

df['Units_sold_MinMax'] = (df['Units_sold']-df['Units_sold'].min())/(df['Units_sold'].max()-df['Units_sold'].min())

print("\nMix-Max Normalized Units_sold")
print(df[['Product','Units_sold','Units_sold_MinMax']])
```

```python
#Z-score normalization :

df['Units_Zscore'] = (df['Units_sold'] - df['Units_sold'].mean())/df['Units_sold'].std()

print("\nZ-score Normalized Units_sold")
print(df[['Product', 'Units_sold', 'Units_Zscore']])
```

###Decimal scaling

```python
#Decimal scaling for price :

df['Price_Decimal'] = df['Price']/1000000

print("\nDecimal Scaled Price")
print(df[['Product', 'Price', 'Price_Decimal']])
```

```python
#Decimal scaling for discount :

df['Discount_Decimal'] = df['Discount']/100

print("\nDecimal Scaled Discount")
print(df[['Product', 'Discount', 'Discount_Decimal']])
```

```python
#Normalize multiple columns

cols = ['Price', 'Units_sold', 'Discount']
df_norm = (df[cols]-df[cols].min()) / (df[cols].max()-df[cols].min())

print("\nNormalized Dataset")
print(df_norm)
```

##14B - Turning categorical variables into quantitive variables in python

```python
data = {

'OrderID' : [ '01', '02', '03', '04', '05', '06', '07', '08', '09', '10'],

'Category' : [ 'Electronics', 'Clothing', 'Cosmetics', 'Grocery', 'Clothing', 'Electronics', 'Cosmetics', 'Grocery', 'Grocery', 'Clothing'],

'Payment' : ['UPI', 'COD', 'UPI', 'Netbanking', 'UPI', "COD", "UPI", "UPI", "Netbanking", "UPI"],

'Customer_gender' : ['Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male'],

'City' : ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Delhi', 'Pune', 'Lucknow', 'Pune', 'Delhi', 'Lucknow'],

'Order_value' :[4500, 5000, 900, 1800, 3300, 7860, 2300, 4855, 3500, 6000]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Method 1 - Label encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender_Label'] = le.fit_transform(df['Customer_gender'])

print(df)
```

```python
#Method 2: One-Hot encoding:

df_encoded = pd.get_dummies(df, columns=['Payment'])

print(df_encoded)
```

```python
#Method 3 : Encoding multiple columns

df_mult = pd.get_dummies(df, columns=['Category', 'City'])

print(df_mult)
```

```python
#Method 4 : Dummy encoding (Drop first)

df_dummy = pd.get_dummies(df, columns=['Payment'], drop_first=True)

print(df_dummy)
```
