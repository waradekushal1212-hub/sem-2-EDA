##Preprocessing on Dataset 1

```python
data = {
    "Roll no" : [1,2,3,4,5,6,7,8],
    "Name" : ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Age" : [20, "-", 21, 19, 20, 22, 21, "20"],
    "Marks" : [99, 87, 76, "-", 88, 95, 91, 89],
    "Department" : ["ECE", "cse", "MECH", "CIVIL", "cse", "ECE", "MECH", "CIVIL"],
    "AdmissionDate" : ["20-06-2025", "21-06-2025", "17-06-2025", "03-06-2026", "01-07-2025", "10/06/2025", "05-06-2025", "28-06-2025"]

}
df = pd.DataFrame(data)

#Save as CSV
df.to_csv("data.csv", index=False)
```

```python
#Replace "-" with NaN to mark missing values

df.replace("-",np.nan,inplace=True)
df
```

```python
#Convert columns to numeric where needed

df["Age"]= pd.to_numeric(df["Age"], errors="coerce")
df["Marks"]= pd.to_numeric(df["Marks"], errors="coerce")
df
```

```python
#Check missing values

print("\nMissing values:")
print(df.isnull().sum())
```

```python
#Fill missing age with mean

df["Age"].fillna(df["Age"].mean(), inplace=True)
df
```

```python
#Fill missing marks with median

df["Marks"].fillna(df["Marks"].median(),inplace=True)
df
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

```python
#Fix incosistent text formatting

df["Department"]= df["Department"].str.upper()
df
```

```python
#Convert admission date to standard datetime format

df["AdmissionDate"]= pd.to_datetime(df["AdmissionDate"], errors= 'coerce')
df
```

```python
#Save cleaned datasheet

df.to_csv("cleaned_student_data.csv", index = False)
print("\nCleaned CSV file saved as cleaned_student_data")
```

##Working  on external dataset with missing values

```python
df=pd.read_csv("/content/Cars93.csv")
df
```

```python
df.isna().sum() #Attributes are displayed
```

###Imputing the missing values

```python
#Luggage Room

df['Luggage.room'] = df['Luggage.room'].fillna(value= df['Luggage.room'].mean())
df
```

```python
df.isna().sum()
```

```python
#Rear seat room

df['Rear.seat.room']= df['Rear.seat.room'].fillna(value= df['Rear.seat.room'].mean())
df
```

```python
df.isna().sum()
```

```python
#Airbags (Categorical data - use mode)

df['AirBags']= df['AirBags'].fillna(value= df['AirBags'].mode().values[0])
df
```

```python
df.isna().sum()
```

```python
#Save cleaned datasheet

df.to_csv("cleaned_cars_data.csv", index = False)
print("\nCleaned CSV file saved as cleaned_cars_data")
```
