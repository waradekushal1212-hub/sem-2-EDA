```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data ={
    'Days' : ['Mon','Tue', 'Wed', 'Thu', 'Fri'],
    'Study_hours' : [2, 2.5,3,4,1.3],
    'Marks' : [75,88,82,90,65] ,
    'Attendance' : [79,98,86,60,55] ,
    'Sleep_hours' : [8,7,7,6,5],
    'Assignments_completed' : [1,3,2,1,2]
}
df = pd.DataFrame(data)
df
```

```python
#1. Line chart (Trend)

plt.plot(df['Days'], df['Study_hours'], marker = 'o')

plt.title('Study hours trend')
plt.xlabel('Days')
plt.ylabel('Study hours')
plt.show()
```

```python
#Advanced line chart

plt.figure(figsize=(10,5))

plt.plot(df['Days'],df['Study_hours'],
         marker='o', color = 'blue', label = 'Study hours')
plt.plot(df['Days'],df['Marks'],
         marker='s', color = 'red', label = 'Marks')

plt.title('Study hours vs Marks trend')
plt.xlabel('Days')
plt.ylabel('Hours/Marks')
plt.legend()
plt
```

```python
#2. Bar chart (comparison)

plt.bar(df['Days'], df['Marks'], color = 'olivedrab')

plt.title('Marks comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
#Advanced bar graph

bars = plt.bar(df['Days'], df['Marks'], color = 'firebrick')

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y,str(y), ha = 'center', va = 'bottom')

plt.title('Marks comparison')
plt.xlabel('Days')
plt.ylabel('Marks')

plt.grid(axis='y') #adds horizontal grid for readability
plt.show()
```

```python
#3. Histogram (Distribtion) - shows how values are spread

data = {
    'Marks' : [75,88,82,90,65,55,34,54,79,56,99,86,76,64,78,54,94,47,86,76,87]
}

df = pd.DataFrame(data)
plt.hist(df['Marks'], bins = 10 , color = 'darkcyan')
plt.title('Marks distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
#4. Scatter plot (Relationship)

import matplotlib.pyplot as plt

plt.scatter(df['Study_hours'], df['Marks'], color = 'navy')

plt.title("Study hours vs Marks")
plt.xlabel("Study hours")
plt.ylabel("Marks")
plt.show()
```

```python
#Conditional scatter plot - example category

df['Result'] = ['Fail','Pass', 'Fail','Pass', 'Pass']
colors = ['red' if r=='Fail' else 'green' for r in df['Result']]

plt.scatter(df['Study_hours'], df['Marks'], c = colors)

plt.title("Study hours vs Marks (result-wise)")
plt.xlabel("Study hours")
plt.ylabel("Marks")
plt.show()
```

##Same graphs using Seaborn (better visualization)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


#Dataset 2

data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri','Sat', 'Sun', 'Mon', 'Tue', 'Wed'],
    'Region' : ['North','South','North', 'East','West', 'East', 'South', 'East', 'West','North'],
    'Sales' : [200,280,380,234,540,455,490,345, 295,410],
    'Profit' : [20,30,35,22,45,55,50,38,40,42],
    'Customers' : [20,45,80,105,115,56,82,90,65,75],
    'Category' : ['Food', 'Clothing', 'Electronics','Clothing', 'Food','Food','Food','Electronics','Electronics','Clothing' ]

}

df = pd.DataFrame(data)
print(df)
```

```python
#Line plot

sns.lineplot(x = 'Day', y = 'Sales', data=df)
plt.title('Seaborn line plot')
plt.show()
```

```python
#Bar plot

sns.barplot(x = 'Day', y = 'Sales', data=df)
plt.title('Seaborn bar plot')
plt.show()
```

```python
#Histogram

sns.histplot(x = 'Sales', data=df, bins = 8)
plt.title('Seaborn histogram')
plt.show()
```

```python
#Scatter plot

sns.scatterplot(x = 'Customers', y = 'Profit', data=df)
plt.title('Seaborn scatter plot')
plt.show()
```
